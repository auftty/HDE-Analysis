```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# 1. ОПТИМИЗАЦИЯ: Работа с размером данных
LIMIT_ROWS = 500000
if len(df) > LIMIT_ROWS:
    df_for_corr = df.sample(n=LIMIT_ROWS, random_state=42)
else:
    df_for_corr = df

# 2. РАСЧЕТ КОРРЕЛЯЦИИ
corr_matrix = df_for_corr.corr(numeric_only=True)

# 3. НАСТРОЙКА ЦВЕТА (Blue - White - Blue)
# Создаем градиент: Синий (-1) -> Белый (0) -> Синий (1)
# Используем hex-коды: #00008b (DarkBlue) -> #ffffff (White) -> #00008b (DarkBlue)
custom_cmap = LinearSegmentedColormap.from_list("custom_blue", ["#003399", "white", "#003399"])

# 4. ВИЗУАЛИЗАЦИЯ
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
plt.figure(figsize=(12, 10))

sns.heatmap(
    corr_matrix, 
    mask=mask, 
    annot=True, 
    fmt=".2f", 
    cmap=custom_cmap,    # Применяем наш кастомный сине-бело-синий градиент
    center=0,            # Гарантируем, что 0 будет ровно посередине (белым)
    vmin=-1, vmax=1,     # Фиксируем границы
    linewidths=0.5, 
    square=True,
    cbar_kws={"shrink": 0.7}
)

plt.title('Матрица корреляций (Сила связи)', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.show()
```
