````python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import ticker as mtick
import pandas as pd
import matplotlib.font_manager as fm
import os
from matplotlib.patches import Patch

BRAND_COLORS = {
    'primary': '#008C44',
    'accent': '#E21E26', 
    'text': '#333333',
    'grid': '#EEEEEE',
    'border': '#CCCCCC',
    'white': '#FFFFFF'
}

def set_chart_design(font_path=None):
    font_name = 'Arial'
    
    if font_path and os.path.exists(font_path):
        fm.fontManager.addfont(font_path)
        prop = fm.FontProperties(fname=font_path)
        font_name = prop.get_name()

    sns.set_style("whitegrid")
    
    plt.rcParams.update({
        'font.family': 'sans-serif',
        'font.sans-serif': [font_name, 'Arial', 'DejaVu Sans'],
        'text.color': BRAND_COLORS['text'],
        'axes.labelcolor': BRAND_COLORS['text'],
        'xtick.color': BRAND_COLORS['text'],
        'ytick.color': BRAND_COLORS['text'],
        'axes.edgecolor': BRAND_COLORS['border'],
        'grid.color': BRAND_COLORS['grid'],
        'figure.facecolor': BRAND_COLORS['white'],
        'axes.facecolor': BRAND_COLORS['white'],
        'axes.spines.top': False,
        'axes.spines.right': False,
        'figure.dpi': 120,
        'axes.titlelocation': 'left',
        'axes.titleweight': 'bold',
        'axes.titlesize': 14,
        'axes.titlepad': 20,
    })

FONT_PATH = "/kaggle/input/dano2025-voronovofilesfromgithub/OTF/5kaSansDesign-Regular.otf"
set_chart_design(FONT_PATH)

col_check = 'Средний чек'
col_trafic = 'Трафик пеший, в час'
col_schools = 'Школы (300 м)'
col_stops = 'Остановки (300 м)'
col_shops = 'Продуктовые магазины (500 м)'
col_marketplaces = 'Маркетплейсы, доставки, постаматы (100 м)'
col_med = 'Медицинские уч. и аптеки (300 м)'

df['has_school'] = df[col_schools] > 0
df['has_stop'] = df[col_stops] > 0
df['has_shop'] = df[col_shops] > 0
df['has_marketplace'] = df[col_marketplaces] > 0
df['has_med'] = df[col_med] > 0

groups = {
    "Школа": df.loc[df['has_school'], col_trafic].mean() / df.loc[~df['has_school'], col_trafic].mean() - 1,
    "Продуктовый магазин": df.loc[df['has_shop'], col_trafic].mean() / df.loc[~df['has_shop'], col_trafic].mean() - 1,
    "Маркетплейс/постамат": df.loc[df['has_marketplace'], col_trafic].mean() / df.loc[~df['has_marketplace'], col_trafic].mean() - 1,
    "Аптека/мед. учреждение": df.loc[df['has_med'], col_trafic].mean() / df.loc[~df['has_med'], col_trafic].mean() - 1,
}

plot_df = (
    pd.DataFrame({
        'Объект инфраструктуры': list(groups.keys()),
        'uplift': list(groups.values())
    })
    .assign(uplift_pct=lambda d: d['uplift'] * 100)
    .sort_values('uplift_pct', ascending=False)
    .reset_index(drop=True)
)

colors = [
    BRAND_COLORS['accent'] if name == "Школа" else BRAND_COLORS['primary'] 
    for name in plot_df['Объект инфраструктуры']
]

fig, ax = plt.subplots(figsize=(10, 6))

sns.barplot(
    data=plot_df,
    x='Объект инфраструктуры',
    y='uplift_pct',
    ax=ax,
    palette=colors,
    alpha=0.9,
)

for patch in ax.patches:
    patch.set_hatch('//')
    patch.set_edgecolor('white')
    patch.set_linewidth(0)

ax.set_title('Отклонение пешеходного трафика при наличии объекта инфраструктуры\nот магазинов без данного объекта')
ax.set_xlabel('')
ax.set_ylabel('Отклонение от среднего трафика, %', fontsize=11)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, ha='center')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=100, decimals=0))

legend_elements = [
    Patch(facecolor=BRAND_COLORS['accent'], edgecolor='white', hatch='//', label='Школа'),
    Patch(facecolor=BRAND_COLORS['primary'], edgecolor='white', hatch='//', label='Прочая инфраструктура')
]
ax.legend(handles=legend_elements, loc='upper right', framealpha=0.9)

plt.tight_layout()
plt.show()
````
