Описание структуры данных:
Предоставлена выгрузка информации о магазинах торговой сети «Пятёрочка» за один календарный год в виде переменной Pandas Dataframe, называется df. Данные собраны на уровне отдельных торговых точек и включают как операционные показатели (трафик, средний чек), так и внешние факторы окружения: инфраструктуру, плотность населения, соседние объекты и конкурентную среду. В целях защиты коммерческой тайны отдельные числовые поля были модифицированы. При этом сохранены относительные соотношения между значениями, чтобы данные оставались пригодными для анализа.

1. new_id – Уникальный идентификатор магазина (анонимизированный). Единицы измерения / тип: int. Пример: 8.
2. Месяц – Месяц наблюдения. Единицы измерения / тип: int. Пример: 5.
3. Трафик – Среднее количество посетителей в месяц. Единицы измерения / тип: чел. Пример: 77325.
4. Средний чек – Средний чек покупателя. Единицы измерения / тип: руб. Пример: 637.942.
5. Дата открытия, категориальный – Группа по дате открытия магазина. Единицы измерения / тип: категориальный. Пример: открыт давно.
6. Торговая площадь, категориальный – Категория торговой площади. Единицы измерения / тип: категориальный. Пример: большой.

````python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import ticker as mtick
import pandas as pd

# -------------------------------------------------------------------------
# 1. ЗАДАЕМ НАЗВАНИЯ СТОЛБЦОВ (df уже должен быть в памяти)
# -------------------------------------------------------------------------
col_check = 'Средний чек'
col_trafic = 'Трафик пеший, в час'
col_schools = 'Школы (300 м)'
col_stops = 'Остановки (300 м)'
col_shops = 'Продуктовые магазины (500 м)'
col_marketplaces = 'Маркетплейсы, доставки, постаматы (100 м)'
col_med = 'Медицинские уч. и аптеки (300 м)'

# -------------------------------------------------------------------------
# 2. ПОДГОТОВКА ФЛАГОВ И АГРЕГАЦИЯ
# -------------------------------------------------------------------------
df['has_school'] = df[col_schools] > 0
df['has_stop'] = df[col_stops] > 0
df['has_shop'] = df[col_shops] > 0
df['has_marketplace'] = df[col_marketplaces] > 0
df['has_med'] = df[col_med] > 0

overall_mean = df[col_trafic].mean()

groups = {
    "Школа": df.loc[df['has_school'], col_trafic].mean() / overall_mean - 1,
    "Продуктовый магазин": df.loc[df['has_shop'], col_trafic].mean() / overall_mean - 1,
    "Маркетплейс/постамат": df.loc[df['has_marketplace'], col_trafic].mean() / overall_mean - 1,
    "Аптека/мед. учреждение": df.loc[df['has_med'], col_trafic].mean() / overall_mean - 1,
}

plot_df = (
    pd.DataFrame({
        'Объект инфраструктуры': list(groups.keys()),
        'uplift': list(groups.values())
    })
    .assign(uplift_pct=lambda d: d['uplift'] * 100)
    .sort_values('uplift_pct', ascending=False)
    .reset_index(drop=True)  # Сбрасываем индекс после сортировки
)

# -------------------------------------------------------------------------
# 3. НАСТРОЙКА ВИЗУАЛЬНОГО СТИЛЯ
# -------------------------------------------------------------------------
pyaterochka_green = '#008C44'
accent_red = '#E21E26'
dark_text = '#333333'

sns.set_style("whitegrid")
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial', 'DejaVu Sans'],
    'text.color': dark_text,
    'axes.labelcolor': dark_text,
    'xtick.color': dark_text,
    'ytick.color': dark_text,
    'axes.edgecolor': '#CCCCCC',
    'grid.color': '#EEEEEE',
    'figure.dpi': 120
})

# -------------------------------------------------------------------------
# 4. СОЗДАЕМ СПИСОК ЦВЕТОВ (Школа — красный, остальные — зеленый)
# -------------------------------------------------------------------------
colors = [
    accent_red if name == "Школа" else pyaterochka_green 
    for name in plot_df['Объект инфраструктуры']
]

# -------------------------------------------------------------------------
# 5. ПОСТРОЕНИЕ ГРАФИКА
# -------------------------------------------------------------------------
fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('white')

sns.barplot(
    data=plot_df,
    x='Объект инфраструктуры',
    y='uplift_pct',
    ax=ax,
    palette=colors,  # <-- Используем список цветов вместо одного цвета
    alpha=0.9,
)

# Штриховка с сохранением цветов
for patch, color in zip(ax.patches, colors):
    patch.set_hatch('//')
    patch.set_edgecolor('white')
    patch.set_linewidth(0.5)

# Подписи и оформление
ax.set_title(
    'Отклонение пешеходного трафика при наличии объектов инфраструктуры\nот среднего по выборке',
    fontsize=14,
    fontweight='bold',
    pad=20
)
ax.set_xlabel('', fontsize=11)
ax.set_ylabel('Отклонение от среднего трафика, %', fontsize=11)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, ha='center')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=100, decimals=0))

# -------------------------------------------------------------------------
# 6. ДОБАВЛЯЕМ ЛЕГЕНДУ (опционально)
# -------------------------------------------------------------------------
from matplotlib.patches import Patch

legend_elements = [
    Patch(facecolor=accent_red, edgecolor='white', hatch='//', label='Школа'),
    Patch(facecolor=pyaterochka_green, edgecolor='white', hatch='//', label='Прочая инфраструктура')
]
ax.legend(handles=legend_elements, loc='upper right', framealpha=0.9)

plt.tight_layout()
plt.show()
````
